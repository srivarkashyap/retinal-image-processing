```python
import os
import numpy as np
import cv2
from PIL import Image
from imgaug import augmenters as iaa
from keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
```

```python
# Define paths to input and output directories
input_dir = '/content/drive/MyDrive/DR_CLAHE'
output_dir = '//content/drive/MyDrive/DR_CAUG'
```

```python
# Define paths to input img and output directories
input_image_path = "/content/drive/MyDrive/GAN/img.jpg"
output_dir = "/content/drive/MyDrive/GAN"
```

```python
# Define GAN image data generator
seq1 = iaa.Sequential([
        iaa.Fliplr(p=0.5),
        iaa.Flipud(p=0.5),
        iaa.Affine(scale={"x": (0.8, 1.2), "y": (0.8, 1.2)}),
        iaa.Affine(translate_percent={"x": (-0.2, 0.2), "y": (-0.2, 0.2)}),
        iaa.Affine(shear=(-20, 20))
])

seq2 = iaa.Sequential([
        iaa.Affine(rotate=(-20, 20)),
        iaa.GammaContrast(gamma=(0.8, 1.2)),
        iaa.AdditiveGaussianNoise(scale=(0, 0.05*255)),
        iaa.ElasticTransformation(alpha=50, sigma=5),
        iaa.PerspectiveTransform(scale=(0.01, 0.1))
])
```

```python
# Loop through each subfolder in the root directory
for subfolder in os.listdir(input_dir):
    subfolder_path = os.path.join(input_dir, subfolder)

    # Create a new subfolder to hold the augmented images
    new_subfolder_path = os.path.join(output_dir, subfolder+"_augmented")
    os.makedirs(new_subfolder_path, exist_ok=True)

    # Loop through each image in the subfolder
    for img_name in os.listdir(subfolder_path):
        img_path = os.path.join(subfolder_path, img_name)
        img = cv2.imread(img_path)

        # Generate 5 different augmented images using seq1
        for i in range(5):
            img_aug1 = seq1.augment_image(img)
            img_aug1_path = os.path.join(new_subfolder_path, img_name[:-4] + f'_aug1_{i}.jpg')
            cv2.imwrite(img_aug1_path, img_aug1)

        # Generate 5 different augmented images using seq2
        for i in range(5):
            img_aug2 = seq2.augment_image(img)
            img_aug2_path = os.path.join(new_subfolder_path, img_name[:-4] + f'_aug2_{i}.jpg')
            cv2.imwrite(img_aug2_path, img_aug2)
```

```python
# Define GAN image data generator
seq1 = iaa.Sequential([iaa.Fliplr(p=0.5)])
seq2 = iaa.Sequential([iaa.Flipud(p=0.5)])
seq3 = iaa.Sequential([iaa.Affine(scale={"x": (0.8, 1.2), "y": (0.8, 1.2)})])
seq4 = iaa.Sequential([iaa.Affine(translate_percent={"x": (-0.2, 0.2), "y": (-0.2, 0.2)})])
seq5 = iaa.Sequential([iaa.Affine(shear=(-20, 20))])

seq6 = iaa.Sequential([iaa.Affine(rotate=(-20, 20))])
seq7 = iaa.Sequential([iaa.GammaContrast(gamma=(0.8, 1.2))])
seq8 = iaa.Sequential([iaa.AdditiveGaussianNoise(scale=(0, 0.05*255))])
seq9 = iaa.Sequential([iaa.ElasticTransformation(alpha=50, sigma=5)])
seq10 = iaa.Sequential([iaa.PerspectiveTransform(scale=(0.01, 0.1))])
```

```python
# Load the input image
img = cv2.imread(input_image_path)

img_aug1 = seq1.augment_image(img)
output_path = os.path.join(output_dir, f"Fliplr.jpg")
cv2.imwrite(output_path, img_aug1)

img_aug2 = seq2.augment_image(img)
output_path = os.path.join(output_dir, f"Flipud.jpg")
cv2.imwrite(output_path, img_aug2)

img_aug3 = seq3.augment_image(img)
output_path = os.path.join(output_dir, f"Affine(scale).jpg")
cv2.imwrite(output_path, img_aug3)

img_aug4 = seq4.augment_image(img)
output_path = os.path.join(output_dir, f"Affine(translate_percent).jpg")
cv2.imwrite(output_path, img_aug4)

img_aug5 = seq5.augment_image(img)
output_path = os.path.join(output_dir, f"Affine(shear).jpg")
cv2.imwrite(output_path, img_aug5)

img_aug6 = seq6.augment_image(img)
output_path = os.path.join(output_dir, f"Affine(rotate).jpg")
cv2.imwrite(output_path, img_aug6)

img_aug7 = seq7.augment_image(img)
output_path = os.path.join(output_dir, f"GammaContrast.jpg")
cv2.imwrite(output_path, img_aug7)

img_aug8 = seq8.augment_image(img)
output_path = os.path.join(output_dir, f"AdditiveGaussianNoise.jpg")
cv2.imwrite(output_path, img_aug8)

img_aug9 = seq9.augment_image(img)
output_path = os.path.join(output_dir, f"ElasticTransformation.jpg")
cv2.imwrite(output_path, img_aug9)

img_aug10 = seq10.augment_image(img)
output_path = os.path.join(output_dir, f"PerspectiveTransform.jpg")
cv2.imwrite(output_path, img_aug10)
```

```python
# Load the input image
img = cv2.imread(input_image_path)

# Display the original image
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')
plt.show()

augmentations = [seq1, seq2, seq3, seq4, seq5, seq6, seq7, seq8, seq9, seq10]
augmentation_names = [
    "Fliplr", "Flipud", "Affine(scale)", "Affine(translate_percent)", "Affine(shear)",
    "Affine(rotate)", "GammaContrast", "AdditiveGaussianNoise", "ElasticTransformation", "PerspectiveTransform"
]

for aug, aug_name in zip(augmentations, augmentation_names):
    img_aug = aug.augment_image(img)
    output_path = os.path.join(output_dir, f"{aug_name}.jpg")
    cv2.imwrite(output_path, img_aug)
    print(f"Saved augmented image with {aug_name}: {output_path}")

    # Display the augmented image
    plt.imshow(cv2.cvtColor(img_aug, cv2.COLOR_BGR2RGB))
    plt.title(aug_name)
    plt.axis("off")
    plt.show()
```
